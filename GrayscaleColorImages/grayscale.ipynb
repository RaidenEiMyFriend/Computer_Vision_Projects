```python
import cv2
from google.colab.patches import cv2_imshow
#rgb image of Emma Stone
image = cv2.imread("emma.jpg")
cv2_imshow(image)
```

```python
# load using opencv and convert to grayscale image
image1 = cv2.imread("emma.jpg" , cv2.IMREAD_GRAYSCALE)
cv2_imshow(image1)
```

```python
# Emma Stone in grayscale, we want to convert this grayscale image to an RGB image.
image2 = cv2.imread("grayscaleemma.jpg")
cv2_imshow(image2)
```

```python
import copy
image3 = copy.deepcopy(image2)
```

```python
# we will create RGB channels by multipling channels with factors, we might not get the result we want but this is the only way
# without using deep learning :))
red = image2[:, :, 0] * 2.5
green = image2[:, :, 0] * 2.3
blue = image2[:, :, 0] * 2.8
image3[:, :, 0] = red
image3[:, :, 1] = green
image3[:, :, 2] = blue
```

```python
# show the RGB image
cv2_imshow(image3)
```
